# Reading the iris dataset

```r
library(readxl)
data <- read_excel("irisds.xlsx")
```

# viewing dataset


```r
head(data)
```

# removing the string valued column from the dataset


```r
newdata <-(data[,-5])
```

# Correlation in dataset


```r
cc <- cor(newdata)
cc
```

# From the above we can see that ("petal_width,petal_length") are highly correlated with the value of 0.9627571


# Linear model for the above


```r
pl <- newdata$petal_length
pw <- newdata$petal_width
linear <- lm(pl~pw)
linear
```

# the intercept value is 1.091


# plot 


```r
plot(linear)
```

# model summary


```r
modelsummary <- summary(linear)
modelsummary
```

# Scatter plot


```r
scatter.smooth(x=newdata$petal_length,y=newdata$petal_width,main="Linear Model Scatter Plot")
```
